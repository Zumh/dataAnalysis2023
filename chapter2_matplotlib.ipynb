**Concepts and Examples:**

1. **Importing Matplotlib:**
   - Importing Matplotlib is typically done with the abbreviation `plt`.
   - Example:
     ```python
     import numpy as np
     import matplotlib.pyplot as plt
     ```

2. **Creating a Simple Figure:**
   - You start by having some data to visualize. In this example, an array `a` is used.
   - Example:
     ```python
     a = np.array([2, 5, 7, 4, 7, 0, 3, 1, 9, 2])
     ```

3. **Plotting Data:**
   - The `plot` function is used to create the graph, and it takes the data array as input.
   - Example:
     ```python
     plt.plot(a)  # Plot the points in the array 'a'
     ```

4. **Adding Titles and Labels:**
   - You can enhance the figure by adding titles and labels for the x and y axes.
   - Example:
     ```python
     plt.title("My first figure")  # Add a title to the figure
     plt.xlabel("My x-axis")       # Give a label to the x-axis
     plt.ylabel("My y-axis")       # Give a label to the y-axis
     ```

5. **Displaying the Figure:**
   - The `show` function is used to display the figure.
   - Example:
     ```python
     plt.show()  # Tell Matplotlib to output the figure. Not strictly required in notebooks but neater.
     ```

6. **Figure Components:**
   - A Matplotlib figure consists of key components: `figure` and `axes`.
   - A `figure` can contain one or more `axes`, which are used for plotting.
   - Example:
     - The terminology is shown in an image in the text.

These are the essential concepts and examples for creating a simple Matplotlib figure and customizing it with titles and labels. The `plot` function is used to visualize data, and Matplotlib allows for fine-tuning the appearance of the figure.

**Creating Subfigures with `plt.subplots`**
- `plt.subplots` is used to create a figure with multiple subfigures in matplotlib.
- It takes the number of rows and columns for the subfigures as parameters and returns a figure object and an array of subfigures (axes).

**Example:**
```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 2)  # Creates a 2x2 grid of subfigures
print(ax.shape)  # Prints the shape of the axes array, which is (2, 2)
ax[0, 0].plot(np.arange(6))  # Accessing and plotting on the top-left subfigure
ax[0, 1].plot(np.arange(6, 0, -1))  # Top-right subfigure
ax[1, 0].plot((-1) ** np.arange(6))  # Bottom-left subfigure
ax[1, 1].plot((-1) ** np.arange(1, 7))  # Bottom-right subfigure
plt.show()  # Display the entire figure
```

**Creating Subfigures with `plt.subplot`**
- `plt.subplot` is an alternative way to create subfigures in matplotlib.
- It uses 1-indexing for subplots and allows you to specify the position of each subplot directly.
- You call `plt.subplot` before each plotting command.

**Example:**
```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(2, 2, 1)  # Create a 2x2 grid and select the first subplot
plt.plot(np.arange(6))  # Plot in the first subplot
plt.subplot(2, 2, 2)  # Select the second subplot
plt.plot(np.arange(6, 0, -1))  # Plot in the second subplot
plt.subplot(2, 2, 3)  # Select the third subplot
plt.plot((-1) ** np.arange(6))  # Plot in the third subplot
plt.subplot(2, 2, 4)  # Select the fourth subplot
plt.plot((-1) ** np.arange(1, 7))  # Plot in the fourth subplot
plt.show()  # Display the entire figure
```

**Using `plt` and Axes Objects**
- You can choose between using the global `plt` object and its functions to refer to the current figure and axes or use multiple figure and axes objects and their methods to work with multiple subfigures at the same time.

**Note:**
- The choice is similar to using random number generators in Python. `np.random.randn` uses the global random number generator, while creating separate generators with `np.random.RandomState(seed)` allows you to use multiple generators simultaneously.

This information should help you work with subfigures and subplots effectively in matplotlib, whether you prefer using `plt.subplots` or `plt.subplot`.
